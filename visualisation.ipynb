```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np

sns.set_style("whitegrid")
```

```python
df=pd.read_csv('fsw_clean_dataset.csv')
```

```python
df.hist(figsize=(10,8), bins=30)

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10,6))

sns.boxplot(data=df)

plt.title("Outlier Detection in Dataset")
plt.show()
```

```python
plt.figure(figsize=(8,6))

sns.heatmap(df.corr(), annot=True, cmap="coolwarm")

plt.title("Feature Correlation Heatmap")
plt.show()
```

```python
plt.figure(figsize=(7,5))

plt.scatter(df["HeatInput"], df["Yield"], alpha=0.6)

plt.xlabel("Heat Input")
plt.ylabel("Yield Stress")
plt.title("Yield Stress vs Heat Input")

plt.show()
```

```python
sns.pairplot(df)

plt.show()
```

```python
from mpl_toolkits.mplot3d import Axes3D

fig = plt.figure(figsize=(8,6))
ax = fig.add_subplot(111, projection='3d')

ax.scatter(df["D"], df["V"], df["Yield"])

ax.set_xlabel("Diameter")
ax.set_ylabel("Speed")
ax.set_zlabel("Yield Stress")

plt.show()
```
